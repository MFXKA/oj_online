```python
from PIL import Image
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt 
```

```python
img = Image.open('lena.bmp') 
image_array = np.array(img)
plt.imshow(img)
```

```python
h,w= image_array.shape
print('image height : {0}  \nimage width : {1} '.format(h,w))
```

# <font color=blue>upside-down lena.im<font>
    This means reverse the height but not width
    
    sudo : setup a numpy matrix for new img than fill the black 

```python
img_ud=np.empty((h,w), dtype=float, order='C')
for i in range(h):
    for j in range(w):
        img_ud[w-1-i][j]=image_array[i][j]
plt.subplot(1,2,1) 
plt.imshow(img)
plt.title('normal')
plt.subplot(1,2,2) 
plt.imshow(Image.fromarray(img_ud))
plt.title('upside_down')
```

# <font color=blue>right-side-left lena.im<font>
    This means reverse the width but not height
    
    sudo : setup a numpy matrix for new img than fill the black 

```python
img_rl=np.empty((h,w), dtype=float, order='C')
for i in range(h):
    for j in range(w):
        img_rl[i][h-1-j]=image_array[i][j]
plt.subplot(1,2,1) 
plt.imshow(img)
plt.title('normal')
plt.subplot(1,2,2) 
plt.imshow(Image.fromarray(img_rl))
plt.title('right-side-left')        
                
```

# <font color=blue>diagonally mirrored lena.im<font>
    This means reverse both width and height
    sudo : setup a numpy matrix for new img than fill the black
     two ways of diagonally mirrored

```python
img_dg=np.empty((h,w), dtype=float, order='C')
for i in range(h):
    for j in range(w):
        img_dg[j][i]=image_array[i][j]
plt.subplot(1,2,1) 
plt.imshow(img)
plt.title('normal')
plt.subplot(1,2,2) 
plt.imshow(Image.fromarray(img_dg))
plt.title('diagonally mirrored lama1')        
                
```

```python
img_dg=np.empty((h,w), dtype=float, order='C')
for i in range(h):
    for j in range(w):
        img_dg[h-1-j][w-1-i]=image_array[i][j]
plt.subplot(1,2,1) 
plt.imshow(img)
plt.title('normal')
plt.subplot(1,2,2) 
plt.imshow(Image.fromarray(img_dg))
plt.title('diagonally mirrored lama-1')        
                
```
